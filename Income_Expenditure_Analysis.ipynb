# Data Visualisation: Income and Expenditure 

### 1. To divide annual household income into income brackets 
### 2. To create a pie chart to check the proportion of various brackets 
### 3. To check for correlation between family members and income, expenditure and income (Annual expenditure: add)
### 4. Plot monthly expenditure against emi 
### 5. Income tax calculation

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Ashna\Downloads\Inc_Exp_Data.csv")
df.describe()
```

### Append income bracket column based on condition of annual household income range

```python
income_bracket = []

for row in df['Annual_HH_Income']:
    if 200000 >= row >= 64200:
        income_bracket.append('low income')
        
    if 500000 >= row >= 200000:
        income_bracket.append('low-mid income')

    if 800000 >= row >= 500000:
        income_bracket.append('mid income')
        
    if 1404000 >= row >= 800000:
        income_bracket.append('high income')
        
df['Income_Bracket'] = income_bracket
df
```

### Plot pie chart based on income bracket 

```python
colors = ['pink', 'silver', 'steelblue', 'blue']
df.groupby(['Income_Bracket']).sum().plot(kind='pie', y='Annual_HH_Income', autopct='%1.0f%%', colors=colors, figsize=(7, 7))
```

### Plot pie chart based on education level of highest qualified member 

```python
colors = ['yellow', 'olive', 'lightgreen', 'darkgreen', 'green']
df.groupby(['Highest_Qualified_Member']).sum().plot(kind='pie', y='Annual_HH_Income', autopct='%1.0f%%', colors=colors, figsize=(7, 7))
```

### Calculating annual household expenditure

```python
df['index1'] = df.index
```

```python
df["Annual_HH_Expenditure"] = df['Mthly_HH_Expense'] * 12
df
```

### Correlation between Annual income and annual expenditure 

```python
column_1 = df["Annual_HH_Income"]
column_2 = df["Annual_HH_Expenditure"]
correlation = column_1.corr(column_2)
print("The correlation between annual income and expenditure:", correlation)

sns.regplot(x=df["Annual_HH_Income"], y=df["Annual_HH_Expenditure"])
```

### Correlation between household income and family members 

```python
column_1 = df["Annual_HH_Income"]
column_2 = df["No_of_Fly_Members"]
correlation = column_1.corr(column_2)
print("The correlation between annual income and family size:", correlation)

sns.regplot(x=df["Annual_HH_Income"], y=df["No_of_Fly_Members"])
```

### Proportion of expenditure spent on emi and rent

```python
ax = df.plot(x="index1", y="Annual_HH_Expenditure", kind="bar")
df.plot(x="index1", y="Emi_or_Rent_Amt", kind="bar", ax=ax, color="orange", figsize = (10, 10))
```

### Income tax calculation


```python
income_tax = []

for row in df['Annual_HH_Income']:
    if 200000 >= row >= 64200:
        income_tax.append(row * 0.1)
        
    if 500000 >= row >= 200000:
        income_tax.append(row * 0.2)
        
    if 800000 >= row >= 500000:
        income_tax.append(row * 0.25)
        
    if 1404000 >= row >= 800000:
        income_tax.append(row * 0.3)
        
df['Income_Tax'] = income_tax
df
```

### Plotted on a line graph

```python
sns.lineplot(data=df, x="Income_Tax", y="Annual_HH_Income")
```
